```python
import pandas as pd

dados = pd.read_csv('teste1.csv', names=['x', 'y1', 'y2', 'y3', 'y4', 'y5', 'y6', 'y7', 'y8'])

entrada = dados.x
saida = dados.y1
```

image.png

```python
n = len(entrada)

sumXY = sum(entrada*saida)
sumX = sum(entrada)
sumY = sum(saida)
sumXX = sum(entrada*entrada)
sumX2 = sumX ** 2

w1 = (n * sumXY - sumX * sumY) / (n * sumXX - sumX2)
print(w1)
```

```python
xm = entrada.mean()
ym = saida.mean()

b = ym - w1 * xm
print(b)
```

image.png

```python
import matplotlib.pyplot as plt

saida_regressao = entrada * w1 + b

plt.plot(entrada, saida)  # .scatter() plota os pontos
plt.plot(entrada, saida_regressao, c='r')
plt.show()
```
